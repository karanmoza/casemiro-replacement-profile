# 04 Visualisations

Generate the bar chart, heatmap, radar chart, scatter plot, and sensitivity chart.

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.append(str(PROJECT_ROOT))

from src.config import CHART_DIR
from src.load_fbref import load_fbref_player_stats
from src.scoring import build_scoring_table, make_shortlist, make_category_scores, sensitivity_analysis
from src.visualisations import save_ranked_score_bar, save_category_heatmap, save_radar_chart, save_security_defence_scatter, save_sensitivity_chart

CHART_DIR.mkdir(parents=True, exist_ok=True)
scored = build_scoring_table(load_fbref_player_stats())
shortlist = make_shortlist(scored)
category_scores = make_category_scores(scored)
sensitivity = sensitivity_analysis(scored)

save_ranked_score_bar(shortlist, CHART_DIR / 'ranked_control_midfielder_score.png')
save_category_heatmap(category_scores, CHART_DIR / 'category_heatmap.png')
save_radar_chart(scored, CHART_DIR / 'casemiro_candidate_radar.png')
save_security_defence_scatter(scored, CHART_DIR / 'security_vs_defensive_floor.png')
save_sensitivity_chart(sensitivity, CHART_DIR / 'sensitivity_analysis.png')
```
